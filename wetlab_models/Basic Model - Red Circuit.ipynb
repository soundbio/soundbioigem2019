```python
model red()
    J0: -> S1 ; v_1 #production of cph8
    J1: S1 -> ; v_2 * S1 #degradation of cph8
    J2: -> S2 ; v_3 * S3 #production of dCBD-sfGFP
    J3: S2 -> ; v_4 * S2 #degradation of dCBD-sfGFP
    J4: $X0 -> S3 ; X0 * S1 * v_5 #phosphorylation of ompR
    J5: S3 -> ; S3 * v_6 #degradation of phosphorylated ompR
    
    # Parameters
    v_1 = 1 
    v_2 = 0.3
    v_3 = 0.5
    v_4 = 0.2
    v_5 = 0.2
    v_6 = 0.9
    
    # Initial values
    X0 = 100 #unphosphorylated ompR
    S1 = 0 #cph8
    S2 = 0 #dCBD-sfGFP
    S3 = 0 #phosphorylated ompR
end
```

```python
import tellurium as telly
red.reset()
red.simulate(0, 1000, 10)
red.plot()
print(red.steadyState())
print(red.S1/red.S2)
print(red.v_4/red.v_3)
```
